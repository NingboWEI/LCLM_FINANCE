## step1. get company tickers id numers

- easy to access through `https://www.sec.gov/files/company_tickers.json`
- no need to sign in/sign up, but have to set `User-Agent`, which is project name+email


```python
import requests
import json

def download_company_tickers(save_path="company_tickers.json"):
    url = "https://www.sec.gov/files/company_tickers.json"
    
    headers = {
        "User-Agent": "LCLM_FINANCE ningbowei56@gmail.com"
    }

    try:
        response = requests.get(url, headers=headers)
        response.raise_for_status()

        data = response.json()

        with open(save_path, "w", encoding="utf-8") as f:
            json.dump(data, f, indent=4)

        print(f"Saved {len(data)} companies to {save_path}")

    except Exception as e:
        print("Download failed:", e)
```

```python
download_company_tickers()
```
